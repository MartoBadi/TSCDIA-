```python
def es_primo(n):
    if n <= 1:
        return False
    contador = 0
    for i in range(2, n):
        if n % i == 0:
            contador += 1
    if contador == 0:
        return True
    else:
        return False 
```

```python
es_primo(2)
```

```python
es_primo(4)
```

```python
def mostrar_primos():
    for i in range(1,101):
        if es_primo(i) == True:
            print(i)

mostrar_primos()
```

¿Se modifica afuera de la función?
¡Sí! La lista original cambia.

¿Por qué?
Porque las listas en Python son objetos mutables. Cuando pasas una lista a una función, se pasa una referencia al mismo objeto, no una copia. Por eso, los cambios dentro de la función afectan a la lista original.

```python
def agregar_elemento(lista, valor):
    lista_2 = lista.copy()
    lista_2.append(valor)
    return lista_2

lista = [1,2]

agregar_elemento(lista, 2)

print(f"La lista con el elemento agregado es {agregar_elemento(lista, 2)}")

print(f"La lista original es {lista}")
```

```python
def promediar():
    total = 0
    contador = 0
    numero_string = 0
    while numero_string != "fin":
        try:
            numero_string = input("Ingrese un número o la palabra 'fin'")
            numero_float = float(numero_string)
            total += numero_float
            contador += 1
        except ValueError: 
            if numero_string == "fin":
                break
            else:
                print("Entrada incorrecta")

    promedio = total/contador

    print(f"el promedio de los números ingresados es: {promedio}")

promediar()
```

```python
def filtrar_positivos(lista):
    copia_de_lista = lista.copy()
    lista_positivos = []
    for i in copia_de_lista:
        if i > 0:
            lista_positivos.append(i)
    return lista_positivos

lista = [1,-2]

print(filtrar_positivos(lista))

print(f"La lista es: {lista}")
```

```python
def pedir_entero(mensaje):
    entero = 0.5
    while type(entero) != int:
        try:
            entero = int(input(mensaje))
        except ValueError:
            pass

    print(f"El número entero ingresado es: {entero}")  

pedir_entero("Ingrese un número entero")
```

```python
def agregar_numero_a_sublistas(lista_de_listas, numero):
    for i in range(len(lista_de_listas)):
        lista_de_listas[i].append(numero)
    return lista_de_listas

agregar_numero_a_sublistas([[1,2],[3,4]],5)
```

[:] crea una copia superficial (shallow copy) de una lista. Solo copia la lista principal, pero si hay objetos mutables dentro (como otras listas), estos no se copian, solo se referencian.
copy.deepcopy() crea una copia profunda (deep copy). Copia la lista y todos los objetos internos, recursivamente. Así, modificar elementos internos en la copia no afecta al original.

```python
def contar_vocales(texto):
    diccionario = {}
    contador_de_as = 0
    contador_de_es = 0
    contador_de_is = 0
    contador_de_os = 0
    contador_de_us = 0
    for letra in texto:
        if (letra == "a") | (letra == "A"):
            contador_de_as += 1
    for letra in texto:
        if (letra == "e") | (letra == "E"):
            contador_de_es += 1
    for letra in texto:
        if (letra == "o") | (letra == "O"):
            contador_de_os += 1
    for letra in texto:
        if (letra == "i") | (letra == "I"):
            contador_de_is += 1
    for letra in texto:
        if (letra == "u") | (letra == "U"):
            contador_de_us += 1
    diccionario["a"] = contador_de_as
    diccionario["e"] = contador_de_es
    diccionario["i"] = contador_de_is
    diccionario["o"] = contador_de_os
    diccionario["u"] = contador_de_us

    return diccionario

contar_vocales("Hola Mundo")
```
