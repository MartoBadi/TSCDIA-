## Resolución

Crea una agenda vacía

```python
agenda = {}
```

En la siguiente celda, incluye el código para ejecutar las siguientes instrucciones:
- Mostrar en pantalla las opciones
- Pedirle al usuario que elija las opciones
- Crear una estructura de control (*if*, *elif* y *else*) que ejecute diferentes bloques de código según las opciones elegidas.

```python

opcion = int(input(""" Elija una opción:
      
1. **Añadir un contacto**
      
2. **Buscar un contacto**:
       
3. **Editar un contacto**: 
      
4. **Eliminar un contacto**:
       
5. **Mostrar todos los contactos**: 
"""))

if opcion == 1:
    nombre = input("Nombre del contacto: ")
    telefono = input("Teléfono del contacto: ")
    agenda[nombre] = telefono
    print(f"Contacto {nombre} añadido con el número {telefono}.")
elif opcion == 2:
    nombre = input("Nombre del contacto a buscar: ")
    if nombre in agenda:
        print(f"El número de {nombre} es {agenda[nombre]}.")
    else:
        print(f"No se encontró el contacto {nombre}.")
elif opcion == 3:
    nombre = input("Nombre del contacto a editar: ")
    if nombre in agenda:
        nuevo_telefono = input(f"Nuevo número para {nombre}: ")
        agenda[nombre] = nuevo_telefono
        print(f"Contacto {nombre} actualizado con el número {nuevo_telefono}.")
    else:
        print(f"No se encontró el contacto {nombre}.")
elif opcion == 4:
    nombre = input("Nombre del contacto a eliminar: ")
    if nombre in agenda:
        del agenda[nombre]
        print(f"Contacto {nombre} eliminado.")
    else:
        print(f"No se encontró el contacto {nombre}.")
elif opcion == 5:
    if agenda:
        print("Lista de contactos:")
        for nombre, telefono in agenda.items():
            print(f"{nombre}: {telefono}")
    else:
        print("No hay contactos en la agenda.")


```
