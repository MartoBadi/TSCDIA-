```python
import nbformat

notebooks = [f'clase_{i}.ipynb' for i in range(1, 18)]

with open(notebooks[0]) as f:
    merged = nbformat.read(f, as_version=4)

for nb_file in notebooks[1:]:
    with open(nb_file) as f:
        nb = nbformat.read(f, as_version=4)
        merged.cells.extend(nb.cells)

with open('todas_las_clases.ipynb', 'w') as f:
    nbformat.write(merged, f)
```
