```python
import nltk
nltk.download('punkt')
from nltk.tree import *
dp1 = Tree('dp', [Tree('d', ['the']), Tree('np', ['dog'])])
dp2 = Tree('dp', [Tree('d', ['the']), Tree('np', ['cat'])])
vp = Tree('vp', [Tree('v', ['chased']), dp2])
tree = Tree('s', [dp1, vp])
print(tree)
print(tree[0,0,0])
print(tree[0,1,0])
print(tree[1,0,0])
print(tree[1,1,0,0])
print(tree[1,1,1,0])
tree.pretty_print()
```
