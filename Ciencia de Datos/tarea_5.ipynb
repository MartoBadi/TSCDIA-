```python
import pandas as pd

data_2024 = {
    "Nombre": ["V.Janacek", "R.Sousa", "J.Gelabert", "R.N.Valle", "N.de Julián", "D.Sikorova", "B.Fayos", "C.Pérez", "Srta.Relaño R", "R.Ramos"],
    "Carreras": [270, 260, 251, 205, 190, 182, 147, 128, 20, 88],
    "Victorias": [52, 49, 38, 33, 24, 22, 19, 15, 10, 8]
}

df_2024 = pd.DataFrame(data_2024)
```

```python
def predict(alpha: float, beta: float, x_i: float) -> float:
    return beta * x_i + alpha
```

```python
def error(alpha: float, beta: float, x_i: float, y_i: float) -> float:
    return predict(alpha, beta, x_i)-y_i
```

```python
import numpy as np

def sum_of_sqerrors(alpha: float, beta: float, x: np.array, y: np.array) -> float:
    return sum(error(alpha, beta, x_i, y_i) ** 2
        for x_i, y_i in zip(x, y))
```

```python
from typing import Tuple

def least_squares_fit(x: np.array, y: np.array) -> Tuple[float, float]:
  beta = np.corrcoef(x, y)[0,1] * np.std(y) / np.std(x)
  alpha = np.mean(y) - beta * np.mean(x)
  return alpha, beta
```

```python
x = [i for i in range(-100, 110, 10)]
y = [3 * i - 5 for i in x]
# Debería hallar que y = 3x–5
assert least_squares_fit(x, y) == (-5, 3)
```

```python
# Calcular alpha y beta
alpha, beta = least_squares_fit(df_2024['Carreras'], df_2024['Victorias'])

# Verificar los resultados
print(f"Alpha: {alpha}, Beta: {beta}")
```

```python
def total_sum_of_squares(y: np.array) -> float:
    return np.sum((y - np.mean(y)) ** 2)

def r_squared(alpha: float, beta: float, x: np.array, y: np.array) -> float:
  y_pred = alpha + beta * x
  ss_total = total_sum_of_squares(y)
  ss_res = np.sum((y - y_pred) ** 2)
  return 1.0 - (ss_res / ss_total)

rsq = r_squared(alpha, beta, df_2024['Carreras'], df_2024['Victorias'])
print(rsq)
```

```python
import random
import tqdm
import numpy as np

num_epochs = 10000
random.seed(0)
guess = [random.random(),
random.random()]
# selecciona valor aleatorio para empezar
learning_rate = 0.00001

def gradient_step(params, gradients, learning_rate):
    params = np.array(params)
    gradients = np.array(gradients)
    return params - learning_rate * gradients

with tqdm.trange(num_epochs) as t:
  for _ in t:
    alpha, beta = guess
# Derivada parcial de pérdida con respecto a alpha
    grad_a = sum(2 * error(alpha, beta, x_i, y_i)
    for x_i, y_i in zip(df_2024['Carreras'],
    df_2024['Victorias']))

# Derivada parcial de pérdida con respecto a beta
    grad_b = sum(2 * error(alpha, beta, x_i, y_i) * x_i
    for x_i, y_i in zip(df_2024['Carreras'],
    df_2024['Victorias']))

# Calcula pérdida para mantener la descripción tqdm
    loss = sum_of_sqerrors(alpha, beta,
    df_2024['Carreras'], df_2024['Victorias'])
    t.set_description(f"loss: {loss:.3f}")

# Por último, actualiza la conjetura
    guess = gradient_step(guess, [grad_a, grad_b], -learning_rate)

# Deberíamos obtener resultados similares:
alpha, beta = guess

print(f"alpha es: {alpha} y beta es: {beta}")
```
