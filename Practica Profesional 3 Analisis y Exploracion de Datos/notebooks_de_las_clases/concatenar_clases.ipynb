```python
import nbformat

# Lista de notebooks a unir
notebooks = ['clase_1.ipynb', 'clase_2.ipynb', 'clase_3.ipynb', 'clase_4.ipynb', 
            'clase_5.ipynb', 'clase_6.ipynb', 'clase_7.ipynb']

# Cargar el primero como base
with open(notebooks[0]) as f:
    merged = nbformat.read(f, as_version=4)

# Aadir las celdas de los dems
for nb_file in notebooks[1:]:
    with open(nb_file) as f:
        nb = nbformat.read(f, as_version=4)
        merged.cells.extend(nb.cells)

# Guardar el notebook concatenado
with open('todas_las_clases.ipynb', 'w') as f:
    nbformat.write(merged, f)
```
