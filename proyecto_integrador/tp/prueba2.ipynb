### <font color="#004D7F"> Olist Brazilian E-Commerce</font>


El dataset Olist Brazilian E-Commerce ofrece una visiÃ³n integral del comercio electrÃ³nico en Brasil entre 2017 y 2018, con mÃ¡s de 99,000 pedidos realizados en mÃºltiples plataformas. Su riqueza radica en la variedad de dimensiones que permite analizar: desde estado del pedido, precios, mÃ©todos de pago, tiempos y costos de envÃ­o, hasta la localizaciÃ³n geogrÃ¡fica de clientes y vendedores, ademÃ¡s de reseÃ±as de clientes y atributos de productos.

Gracias a su estructura, conformada por nueve tablas interrelacionadas con un total de 40 columnas, este dataset es ideal para estudiar el comportamiento del cliente, el rendimiento logÃ­stico, la fidelidad, y la satisfacciÃ³n postventa, ofreciendo oportunidades tanto para anÃ¡lisis descriptivo como predictivo. AdemÃ¡s, la presencia de geolocalizaciÃ³n (latitud y longitud por cÃ³digo postal) permite ampliar los estudios hacia anÃ¡lisis espaciales de mercado.

```python

def load_olist_datasets(path: str) -> dict:
    """
    Lee todos los archivos CSV en la carpeta especificada y devuelve
    un diccionario de DataFrames con el nombre del archivo como clave.
  
    """
    datasets = {}
    
    # Iterar sobre los archivos de la carpeta
    for file in os.listdir(path):
        if file.endswith(".csv"):
            file_path = os.path.join(path, file)
            
            # Nombre del dataset sin extensiÃ³n
            dataset_name = os.path.splitext(file)[0]
            
            # Leer CSV con pandas
            datasets[dataset_name] = pd.read_csv(file_path, encoding="utf-8")
            print(f"âœ… Cargado: {dataset_name} ({datasets[dataset_name].shape[0]} filas, {datasets[dataset_name].shape[1]} columnas)")
    
    return datasets

#Cambiar Ruta
carpeta = r"C:\Users\jazmin\Documents\Proyectos\25_forecast\data\csv"
olist_data = load_olist_datasets(carpeta)

# Acceso a cada dataset
customers = olist_data["olist_customers_dataset"]
orders = olist_data["olist_orders_dataset"]
order_items = olist_data["olist_order_items_dataset"]
payments = olist_data["olist_order_payments_dataset"]
reviews = olist_data["olist_order_reviews_dataset"]
products = olist_data["olist_products_dataset"]
sellers = olist_data["olist_sellers_dataset"]
geolocation = olist_data["olist_geolocation_dataset"]
categories = olist_data["product_category_name_translation"]
```

## <font color="#004D7F"> 1.2 RelaciÃ³n Datasets</font>

```python

# FunciÃ³n para detectar y unificar relaciones entre datasets
def unify_using_relationships(datasets: dict, primary_table: str = "olist_orders_dataset") -> pd.DataFrame:
    """
    Unifica dinÃ¡micamente todos los datasets de Olist usando las relaciones detectadas.
   
    """
    # Copiamos para no modificar los originales
    data = {k: v.copy() for k, v in datasets.items()}

    # 1ï¸âƒ£ Detectar relaciones
    def detect_relationships(datasets):
        relationships = []
        for name1, df1 in datasets.items():
            for name2, df2 in datasets.items():
                if name1 != name2:
                    common_cols = set(df1.columns).intersection(df2.columns)
                    for col in common_cols:
                        if (name2, col, name1, col) not in relationships:
                            relationships.append((name1, col, name2, col))
        return relationships

    relationships = detect_relationships(data)
    print("\nðŸ”— Relaciones detectadas:")
    for r in relationships:
        print(r)

    # 2ï¸âƒ£ Unica tablas partiendo de la tabla primaria
    merged_df = data[primary_table].copy()
    merged_tables = {primary_table}

    for table_a, col_a, table_b, col_b in relationships:
        if table_a in merged_tables and table_b not in merged_tables:
            merged_df = merged_df.merge(
                data[table_b],
                left_on=col_a,
                right_on=col_b,
                how="left",
                suffixes=("", f"_{table_b[:6]}")
            )
            merged_tables.add(table_b)

        elif table_b in merged_tables and table_a not in merged_tables:
            merged_df = merged_df.merge(
                data[table_a],
                left_on=col_b,
                right_on=col_a,
                how="left",
                suffixes=("", f"_{table_a[:6]}")
            )
            merged_tables.add(table_a)

    print(f"\nâœ… Data unificada: {merged_df.shape[0]} filas, {merged_df.shape[1]} columnas")
    return merged_df, relationships


# --- Ejecutar unificaciÃ³n ---
df_master, relationships = unify_using_relationships(olist_data, primary_table="olist_orders_dataset")


# FunciÃ³n para graficar mapa de relaciones (ERD simplificado)
def plot_relationships(relationships):
    """
    Crea un grafo de relaciones entre tablas usando networkx.
    """
    G = nx.Graph()

    # Crear nodos y aristas
    for tabla1, col1, tabla2, col2 in relationships:
        G.add_node(tabla1)
        G.add_node(tabla2)
        G.add_edge(tabla1, tabla2, label=col1)

    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(G, seed=42)
    nx.draw(G, pos, with_labels=True, node_size=2500, node_color="lightblue",
            font_size=10, font_weight="bold", edge_color="gray")
    labels = nx.get_edge_attributes(G, "label")
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=8)
    plt.title("Mapa de Relaciones entre Tablas (ERD Simplificado)", fontsize=14)
    plt.show()

# Graficar relaciones
plot_relationships(relationships)
```

```python
# ==============================================================================
# PASO 3: LIMPIEZA Y PREPROCESAMIENTO
# ==============================================================================
print("\n--- Iniciando limpieza y preprocesamiento ---")

# --- 3.1. Manejo de Infinitos y Nulos ---

reviews['review_comment_message'] = reviews['review_comment_message'].fillna('Sin Comentario')
reviews['review_comment_title'] = reviews['review_comment_title'].fillna('Sin TÃ­tulo')

products["product_category_name"] = products["product_category_name"].fillna("Sin Definir")
products['product_weight_g'] = products.groupby('product_description_lenght')['product_weight_g'].transform(lambda x: x.fillna(x.median()))
products['product_weight_g'] = products['product_weight_g'].fillna(products['product_weight_g'].median())
products.dropna(subset=['product_description_lenght'], inplace=True)
print("Manejo de nulos completado.")

# --- 3.2. ConversiÃ³n de Tipos de Datos Fechas ---
date_columns = [
    'order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date',
    'order_delivered_customer_date', 'order_estimated_delivery_date'
]
for col in date_columns:
    orders[col] = pd.to_datetime(orders[col], errors='coerce')


date_columns = [ 'review_creation_date', 'review_answer_timestamp']
for col in date_columns:
    reviews[col] = pd.to_datetime(reviews[col], errors='coerce')

print("ConversiÃ³n de tipos de datos a datetime completada.")
```

```python
df_filtrado = order_items.merge(orders[['order_id', 'order_purchase_timestamp']], on='order_id', how='left')
df_filtrado = df_filtrado[['order_purchase_timestamp', 'ingreso_total']]
df_filtrado['date_purchase'] = df_filtrado['order_purchase_timestamp'].dt.date
df_sells = df_filtrado.groupby('date_purchase').agg({'ingreso_total': 'sum'}).reset_index()
df_sells.sort_values(by='date_purchase', inplace=True)
df_sells = df_sells.rename(columns={'date_purchase': 'ds', 'ingreso_total': 'y'})
df_sells .head(20)
```

```python


import os
os.environ['TMPDIR'] = r'C:\\Temp'   # o cualquier carpeta donde tengas permisos
os.environ['TEMP'] = r'C:\\Temp'
os.environ['TMP'] = r'C:\\Temp'

# Crear y ajustar el modelo
model = Prophet(
    yearly_seasonality=True,   # Detecta patrones anuales
    weekly_seasonality=True,   # Detecta patrones semanales
    daily_seasonality=False    # Desactiva patrones diarios si no son necesarios
)
model.fit(df_sells)

#  Crear fechas futuras
future = model.make_future_dataframe(periods=120)

#  Hacer predicciones
forecast = model.predict(future)

# Graficar predicciones
fig1 = model.plot(forecast)
plt.title('PredicciÃ³n de Ventas con Prophet')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.show()

#  Graficar los componentes de la predicciÃ³n
fig2 = model.plot_components(forecast)
plt.show()
```

**CÃ¡lculos logÃ­sticos** 
Estos cÃ¡lculos permiten detectar cuellos de botella y medir desempeÃ±o:

| **CÃ¡lculo**                   | **FÃ³rmula**                                                     |
| ----------------------------- | --------------------------------------------------------------- |
| Tiempo de aprobaciÃ³n          | `order_approved_at - order_purchase_timestamp`                  |
| Tiempo hasta transportista    | `order_delivered_carrier_date - order_approved_at`              |
| Tiempo de transporte          | `order_delivered_customer_date - order_delivered_carrier_date`  |
| AnticipaciÃ³n estimada vs real | `order_estimated_delivery_date - order_delivered_customer_date` |
| Lead time total               | `order_delivered_customer_date - order_purchase_timestamp`      |
| Tiempo lÃ­mite de envÃ­o        | `shipping_limit_date - order_purchase_timestamp`                |


```python
# Calcular mÃ©tricas logÃ­sticas
orders['approval_time'] = (orders['order_approved_at'] - orders['order_purchase_timestamp']).dt.total_seconds() / 3600
orders['carrier_dispatch_time'] = (orders['order_delivered_carrier_date'] - orders['order_approved_at']).dt.total_seconds() / 3600
orders['transport_time'] = (orders['order_delivered_customer_date'] - orders['order_delivered_carrier_date']).dt.total_seconds() / 3600
orders['delivery_gap'] = (orders['order_estimated_delivery_date'] - orders['order_delivered_customer_date']).dt.days
orders['lead_time'] = (orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']).dt.days

```
