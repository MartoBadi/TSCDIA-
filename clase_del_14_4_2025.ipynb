### Ejemplo con un vector

```python
# Importamos la librería Numpy
import numpy as np
```

```python
# Generamos un vector aleatorio con media 20 y desviación estándar 10

vec = np.random.normal(20, 10, 10) # El 3er atributo indica la cantidad de elementos del vector

np.mean(vec) # Este es el método para calcular la media utilizando Numpy
```

```python
# Generamos un vector aleatorio que representará un outlier con media 300 y desviación estándar 100

vec_2 = np.random.normal(300, 100, 1)
```

```python
# Añadimos el valor atípico (ruido/outlier) al 1er vector creado

vec_ruid = np.append(vec, vec_2)
```

```python
np.mean(vec_ruid)
```

## Media Truncada

```python
from scipy.stats import trim_mean # importamos el método trim_mean
```

### Ejemplo con un vector

```python
# A través de la media truncada no considero el 10% (en este caso) superior e inferior del conjunto de datos analizados

mean_trimmed_vec = trim_mean(vec, 0.1)
mean_trimmed_vec_ruid = trim_mean(vec_ruid, 0.1)

print(mean_trimmed_vec)
print(mean_trimmed_vec_ruid) # Podemos observar en su salida que es mucho más robusto ante los outliers
```

### Ejemplo con una columna del dataset Iris

```python
mean_trimmed_sepal_length = trim_mean(df['sepal length (cm)'], 0.1)

mean_trimmed_sepal_length
```

## Mediana

### Ejemplo para un vector

```python
median_vec = np.median(vec)
median_vec_ruid = np.median(vec_ruid)

print(median_vec)
print(median_vec_ruid) # Podemos observar en su salida que es mucho más robusto ante los outliers que la media
```

### Ejemplo con una columna del dataset Iris

```python
median_sepal_length = df['sepal length (cm)'].median()

median_sepal_length
```

# Estadísticas de Resumen Multivariadas

## Covarianza

### Con Numpy

```python
covarianza_np = np.cov(df['petal length (cm)'], df['petal width (cm)'])
covarianza_np
```

### Con Pandas

```python
# Obtenemos directamente el valor de covarianza entre los atributos seleccionados

covarianza_pd = df['petal length (cm)'].cov(df['petal width (cm)'])
covarianza_pd
```

```python
# Obtenemos el valor de covarianza en formato DataFrame (excluyendo la columna 'species')

covarianza_pd_df = df.iloc[:,:-1].cov()
covarianza_pd_df
```

```python
type(covarianza_pd_df)
```
